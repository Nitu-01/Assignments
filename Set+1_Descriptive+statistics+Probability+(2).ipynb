# 3) (i)	Where would the mode of this dataset lie?

(ii)	Comment on the skewness of the dataset.	
(iii)	Suppose that the above histogram and the box-plot in question 2 are plotted for the same dataset. Explain how these graphs complement each other in providing information about any dataset.



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming your dataset is stored in a list called 'data'
data = [10, 15, 20, 25, 30, 35, 40, 45, 50]

# Calculate mode
mode = pd.Series(data).mode().values[0]
print("Mode of the Dataset:", mode)

# Calculate skewness
skewness = pd.Series(data).skew()
print("Skewness of the Dataset:", skewness)

# Create a histogram using Seaborn
sns.histplot(data=data, bins='auto', kde=True)
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

# Create a box plot using Seaborn
sns.boxplot(data=data)
plt.title('Box Plot')
plt.show()

```

# 4.	AT&T was running commercials in 1990 aimed at luring back customers who had switched to one of the other long-distance phone service providers. One such commercial shows a businessman trying to reach Phoenix and mistakenly getting Fiji, where a half-naked native on a beach responds incomprehensibly in Polynesian. When asked about this advertisement, AT&T admitted that the portrayed incident did not actually take place but added that this was an enactment of something that “could happen.” Suppose that one in 200 long-distance telephone calls is misdirected. What is the probability that at least one in five attempted telephone calls reaches the wrong number? (Assume independence of attempts.)

```python
import scipy.stats as stats

# Probability of success (getting a wrong number) in a single trial
p = 1 / 200

# Number of trials (attempted telephone calls)
n = 5

# Calculate the probability of getting at least one success
probability_at_least_one_success = 1 - stats.binom.cdf(0, n, p)

print("Probability of at least one wrong number in five calls:", probability_at_least_one_success)
```

# 5.	Returns on a certain business venture, to the nearest
$1,000, are known to follow the following probability distribution
x	P(x)
-2,000	0.1
-1,000	0.1
0	0.2
1000	0.2
2000	0.3
3000	0.1

(i)	What is the most likely monetary outcome of the business venture?
(ii)	Is the venture likely to be successful? Explain
(iii)	What is the long-term average earning of business ventures of this kind? Explain
(iv)	What is the good measure of the risk involved in a venture of this kind? Compute this measure



```python
# Given probability distribution
outcomes = [-2000, -1000, 0, 1000, 2000, 3000]
probabilities = [0.1, 0.1, 0.2, 0.2, 0.3, 0.1]

# (i) Calculate the most likely monetary outcome (mode)
most_likely_outcome = outcomes[probabilities.index(max(probabilities))]
print("Most likely monetary outcome:", most_likely_outcome)

# (ii) Assess the venture's success
# A venture is likely to be successful if the expected value (mean) of the distribution is positive.
expected_value = sum(outcome * probability for outcome, probability in zip(outcomes, probabilities))
is_successful = expected_value > 0
print("Is the venture likely to be successful?", "Yes" if is_successful else "No")

# (iii) Calculate the long-term average earning (expected value)
print("Long-term average earning:", expected_value)

# (iv) Compute the measure of risk (variance)
variance = sum((outcome - expected_value) ** 2 * probability for outcome, probability in zip(outcomes, probabilities))
print("Measure of risk (variance):", variance)
```
